```python
# Convert date and time columns to datetime objects
df['EVENT_LCL_DATE'] = pd.to_datetime(df['EVENT_LCL_DATE'])
df['EVENT_LCL_TIME'] = pd.to_datetime(df['EVENT_LCL_TIME'].str.replace('Z', ''), format='%H:%M:%S', errors='coerce').dt.time

# Combine date and time into a single datetime column
df['EVENT_DATETIME'] = df.apply(lambda row: pd.Timestamp.combine(row['EVENT_LCL_DATE'], row['EVENT_LCL_TIME']) if pd.notna(row['EVENT_LCL_TIME']) else row['EVENT_LCL_DATE'], axis=1)

# Display the updated dataframe with the new datetime column
display(df[['EVENT_LCL_DATE', 'EVENT_LCL_TIME', 'EVENT_DATETIME']].head())
```

```python
df_new['FATAL_FLAG'] = df_new['FATAL_FLAG'].fillna('No')
```

```python
flt_phase_mode = df_new['FLT_PHASE'].mode()[0]
df_new['FLT_PHASE'] = df_new['FLT_PHASE'].fillna(flt_phase_mode)
acft_dmg_desc_mode = df_new['ACFT_DMG_DESC'].mode()[0]
df_new['ACFT_DMG_DESC'] = df_new['ACFT_DMG_DESC'].fillna(acft_dmg_desc_mode)
```

```python
acft_dmg_desc_unknown = []
acft_dmg_desc_minor = []
acft_dmg_desc_substantial = []
acft_dmg_desc_destroyed = []
cols = [acft_dmg_desc_unknown,]
for row in df_new['ACFT_DMG_DESC']:
  if row=='Unknown':
    acft_dmg_desc_unknown.append(1)
    acft_dmg_desc_minor.append(0)
    acft_dmg_desc_substantial.append(0)
    acft_dmg_desc_destroyed.append(0)
  else:
    acft_dmg_desc_unknown.append(0)
    if row=='Minor':
      acft_dmg_desc_minor.append(1)
      acft_dmg_desc_substantial.append(0)
      acft_dmg_desc_destroyed.append(0)
    else:
      acft_dmg_desc_minor.append(0)
      if row=='Substantial':
        acft_dmg_desc_substantial.append(1)
        acft_dmg_desc_destroyed.append(0)
      else:
        acft_dmg_desc_substantial.append(0)
        acft_dmg_desc_destroyed.append(1)
```

```python
df_new['ACFT_DMG_DESC_UNKNOWN'] = pd.Series(acft_dmg_desc_unknown,index=df_new.index)
df_new['ACFT_DMG_DESC_MINOR'] = pd.Series(acft_dmg_desc_minor,index=df_new.index)
df_new['ACFT_DMG_DESC_SUBSTANTIAL'] = pd.Series(acft_dmg_desc_substantial,index=df_new.index)
df_new['ACFT_DMG_DESC_DESTROYED'] = pd.Series(acft_dmg_desc_destroyed,index=df_new.index)
```

```python
flt_phase_text = []
possible_phase = ["LANDING", "TAKEOFF", "CRUISE","APPROACH"]
for i in df_new.index: # Iterate over the indices of df_new
  rmk_text = df_new.loc[i,'RMK_TEXT']
  flag = False
  for phase in possible_phase:
    if phase in rmk_text:
      flt_phase_text.append(phase)
      flag = True
      break
  if not flag:
    flt_phase_text.append("UNKNOWN")
```

```python
df_new['FLT_PHASE_TEXT'] = pd.Series(flt_phase_text,index=df_new.index)
```
