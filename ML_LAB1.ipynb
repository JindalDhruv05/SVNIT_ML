```python
X = df.drop('Sales',axis=1)
y = df['Sales']
```

```python
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train = X_train.T
X_test = X_test.T
```

```python
def lin_reg_model(X: np.ndarray, y: np.ndarray):
  w_size = X.shape[0]
  w = np.zeros(w_size)
  iter = 100000
  grad = np.dot(np.dot(X,X.T),w) - np.dot(X,y)
  while(np.all(grad!=0) or iter>0):
    grad = np.dot(np.dot(X,X.T),w) - np.dot(X,y)
    w = w - 0.005*grad
    iter -= 1
  return w
```

```python
w = lin_reg_model(X_train,y_train)
```

```python
from sklearn.metrics import r2_score

y_pred = np.dot(X_test.T,w)
r2_score(y_test,y_pred)
```
